```python
import pandas as pd
import numpy as np
```

```python
moviesData = pd.read_csv("tmdb_5000_movies.csv")
creditsData = pd.read_csv("tmdb_5000_credits.csv")
# moviesData.head()
creditsData.head()
# moviesData['genres'].head()
# moviesData.shape
```

```python
moviesData = moviesData.merge(creditsData, on="title")
```

```python
moviesData = moviesData[['genres','id','cast', 'crew','overview','title', 'keywords']]
```

```python
moviesData.isnull().sum()
```

```python
moviesData.dropna(inplace=True)
```

```python
import ast
# splits the source code into tokens based on the grammar
```

```python
def helper(obj): # Create a function definition
    Lst = [] # Create an empty list
    for i in ast.literal_eval(obj): # literal_eval used to convert a string of dictionary/json into a dictionary/json
        Lst.append(i['name']) # append the genres names into the list and return the list at the end
    return Lst
```

```python
moviesData['genres'] = moviesData['genres'].apply(helper)
```

```python
moviesData['keywords'] = moviesData['keywords'].apply(helper)
```

```python
moviesData
```

```python
def helperCast(obj): # Create a function definition
    Lst = [] # Create an empty list
    counter = 0
    for i in ast.literal_eval(obj): # literal_eval used to convert a string of dictionary/json into a dictionary/json
        if(counter!=5):
            Lst.append(i['name']) # append the genres names into the list and return the list at the end
            counter=counter+1
        else:
            break
            
    return Lst
```

```python
moviesData['cast'] = moviesData['cast'].apply(helperCast)
```

```python
moviesData.head()
```

```python
def helperCrew(obj): # Create a function definition
    Lst = [] # Create an empty list
    for i in ast.literal_eval(obj): # literal_eval used to convert a string of dictionary/json into a dictionary/json
        if(i['job'] == "Director"):
            Lst.append(i['name']) # append the genres names into the list and return the list at the end
            break
    return Lst
```

```python
moviesData['crew'] = moviesData['crew'].apply(helperCrew)
```

```python
moviesData
```

```python
moviesData.isnull().sum()
```

```python
moviesData
```

```python
moviesData['overview'] = moviesData['overview'].apply(lambda x:x.split())
```

```python
moviesData.head()
```

```python
# We would like to combine the words with spaces together like 'Sam Worthingthon' as 'SamWorthingthon', 'Sam Mendes' as 'SamMendes' 
# so as to avoid confusion for our recommendation system so that it does not recommend Sam Worthingthon's movies for 
# Sam Mendes movies and so on. Same is applied on cast, crew, keywords, genres.'Science Fiction' as 'ScienceFiction' 

moviesData['genres'] = moviesData['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
moviesData['crew'] = moviesData['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
moviesData['cast'] = moviesData['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
moviesData['keywords'] = moviesData['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
moviesData.head()
```

```python
moviesData['tags'] = moviesData['overview'] + moviesData['genres'] + moviesData['keywords'] + moviesData['cast'] + moviesData['crew'] 
```

```python
moviesDf = moviesData[['id','title','tags']]
```

```python
moviesDf['tags'] = moviesDf['tags'].apply(lambda x:" ".join(x))
moviesDf['tags'] = moviesDf['tags'].apply(lambda x:(x.lower()))
```

```python
moviesDf
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 3000, stop_words = 'english')
vector = cv.fit_transform(moviesDf['tags']).toarray()
vector
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def fn(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
moviesDf['tags'].apply(fn)
```

```python
# as number of data increases, the euclidean distance calculation method fails and cosine method is more preffered to find the similarity
# between two vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
moviesDf[moviesDf['title'] == 'Gandhi'].index[0]
```

```python
def recommend(movie):
    index = moviesDf[moviesDf['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(moviesDf.iloc[i[0]].title)
```

```python
recommend('Thor: The Dark World')
```
